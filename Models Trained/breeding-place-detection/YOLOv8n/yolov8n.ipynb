```python
from pathlib import Path

base_path = Path("/kaggle/input/multi-model-for-mosquito-my-fyp/VisText-Mosquito A Multimodal Dataset for Mosquito")
dataset_path = base_path / "Breeding Place Detection"
```

```python
yaml_path = "/kaggle/working/data.yaml"

yaml_content = """
train: /kaggle/input/multi-model-for-mosquito-my-fyp/VisText-Mosquito A Multimodal Dataset for Mosquito/Breeding Place Detection/train/images
val: /kaggle/input/multi-model-for-mosquito-my-fyp/VisText-Mosquito A Multimodal Dataset for Mosquito/Breeding Place Detection/valid/images
test: /kaggle/input/multi-model-for-mosquito-my-fyp/VisText-Mosquito A Multimodal Dataset for Mosquito/Breeding Place Detection/valid/images

nc: 5
names: ["Bottle", "Coconut-Exocarp", "Drain-Inlet", "Tire", "Vase"]

roboflow:
  workspace: adnans-workspace
  project: mosquito-beeding-place
  version: 2
  license: CC BY 4.0
  url: https://universe.roboflow.com/adnans-workspace/mosquito-beeding-place/dataset/2
"""

with open(yaml_path, "w") as f:
    f.write(yaml_content.strip())

print(f" data.yaml created at: {yaml_path}")
```

```python
from ultralytics import YOLO

# Load pre-trained YOLOv8n model
model = YOLO("yolov8n.pt")

# Train the model
results = model.train(
    data=yaml_path,
    epochs=80,
    imgsz=640,
    batch=16,
    name="yolov8n_breeding_detection",
    project="/kaggle/working"
)
```
